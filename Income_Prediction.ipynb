# **DATASET**

```python
df=pd.read_csv("train.csv")
```

## Renaming Columns

```python
df.columns=['Age','Workclass', 'Fnlwgt', 'Education' ,'education_num' ,'marital_status' ,'occupation' ,'relationship' ,'race' ,'sex' ,'capital_gain', 'capital_loss', 'hours_per_week' ,'native_country' ,'income']
df
```

```python
df.isnull().sum()
```

## Data Cleaning

```python
df= df[df.Workclass != ' ?']
df= df[df.occupation != ' ?']
df= df[df.native_country != ' ?']

df
```

## Data Preprocessing

```python
le= lb()

le.fit(df['Workclass'])
lst= le.transform(df['Workclass'])
df['Workclass']= df['Workclass'].replace(to_replace= list(le.classes_), value= list(set(lst)))

#df['Workclass']=le.fit_transform(df['Workclass'])
le.fit(df['Education'])
lst=le.transform(df['Education'])
df['Education']= df['Education'].replace(to_replace= list(le.classes_), value= list(set(lst)))

le.fit(df['marital_status'])
lst=le.transform(df['marital_status'])
df['marital_status']= df['marital_status'].replace(to_replace= list(le.classes_), value= list(set(lst)))

le.fit(df['occupation'])
lst=le.transform(df['occupation'])
df['occupation']= df['occupation'].replace(to_replace= list(le.classes_), value= list(set(lst)))

le.fit(df['relationship'])
lst=le.transform(df['relationship'])
df['relationship']= df['relationship'].replace(to_replace= list(le.classes_), value= list(set(lst)))

le.fit(df['race'])
lst=le.transform(df['race']) #encoding done and stored in same columns
df['race']= df['race'].replace(to_replace= list(le.classes_), value= list(set(lst)))

le.fit(df['sex'])
lst=le.transform(df['sex'])
df['sex']= df['sex'].replace(to_replace= list(le.classes_), value= list(set(lst)))

le.fit(df['native_country'])
lst=le.transform(df['native_country'])
df['native_country']= df['native_country'].replace(to_replace= list(le.classes_), value= list(set(lst)))

le.fit(df['income'])
lst=le.transform(df['income'])
df['income']= df['income'].replace(to_replace= list(le.classes_), value= list(set(lst)))
```

```python
df
```

## Dataset Creation

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size= 0.3)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

**Applying** **Standardization**

```python
scaler= MinMaxScaler()
x_train= scaler.fit_transform(x_train)
x_test= scaler.transform(x_test)
```

# **MODELS**

## A generic functoin for all models

```python
from sklearn.metrics import confusion_matrix as cm,classification_report as cr

def apply_models(m,x_train,x_test,y_train,y_test):
  m.fit(x_train,y_train)
  print('Training Score\n',m.score(x_train,y_train))
  print('\nTesting Score\n',m.score(x_test,y_test))
  y_pred=m.predict(x_test)
  print("\ny_pred\n",y_pred)
  print("\nConfusion Matrix\n",cm(y_test,y_pred))
  print("\nClassification Report\n",cr(y_test,y_pred))
```

## Decision Tree

```python
m_dt= dt(criterion= 'entropy', min_samples_split= 400)
apply_models(m_dt,x_train,x_test,y_train,y_test)
```

## Random Forest Classifier

```python
m_rf= rf(criterion= 'entropy', min_samples_split= 20, max_depth= 20)
apply_models(m_rf,x_train,x_test,y_train,y_test)
```

## Logistic Regression

```python
m_lr= lr()
apply_models(m_lr,x_train,x_test,y_train,y_test)
```

## KNN

```python
m_knn= knn(n_neighbors= 21)
apply_models(m_knn,x_train,x_test,y_train,y_test)
```

## Gaussian Naive Bayes

```python
m_gnb= gnb()
apply_models(m_gnb,x_train,x_test,y_train,y_test)
```
